# Euler tour

A **tour** of a graph $G$ is a ($u, v$)-walk which $u=v$ (**closed walk**) and that traverses each edge in $G$. 

An **Euler tour** is a tour in which all edges are traversed exactly once.
- A connected graph $G$ (with more than one vertex) has an Euler tour if and only if it has no vertices of odd degree.
- A connected graph $G$ has an Euler trail if and only if it has exactly two vertices of odd degree. Moreover, the trail originates and ends in the vertices of odd degree.

**The Chinese postman problem**

Find a closed walk $W = [v_0, e_1, v_1, ..., v_{n-1}, e_n, v_n]$ that covers all edges of $G$, but with minimal weight. Thus, $E(W) = E(G)$ and $\Sigma^n_{i=1} w(e_i)$ is minimal.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

edges = [
    (1, 2), (1, 5),
    (2, 3), (2, 6), (2, 5),
    (3, 4), (3, 7),
    (4, 8),
    (8, 7), (8, 12),
    (9, 5), (9, 10),
    (10, 6),
    (11, 12), (11, 7),
    (5, 6),
    (6, 7),
    (3, 8)
]

G.add_edges_from(edges)

pos = nx.spring_layout(G, seed=4)

nx.draw(G, pos, with_labels=True)
```
